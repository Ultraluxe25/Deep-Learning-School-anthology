```python
class Calculator:
    def __init__(self, x: int, y: int) -> None:
        self.x = x
        self.y = y
        
    def addition(self):
        return self.x + self.y
    
    def subtraction(self):
        return self.x - self.y
    
    def multiplication(self):
        return self.x * self.y
    
    def integer_division(self):
        return self.x // self.y
    
    def exponentation(self):
        return self.x ** self.y
    
    
a, b = map(int, input().split())
my_calculator = Calculator(a, b)

add = my_calculator.addition()
subtract = my_calculator.subtraction()
multiplication = my_calculator.multiplication()
int_div = my_calculator.integer_division()
exp = my_calculator.exponentation()

operations = (add, subtract, multiplication, int_div, exp)
for activity in operations:
    print(activity)
```
