```python
import numpy as np
```

```python
a = np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])
```

```python
elements = [a[0]]
print(elements)
freq = [1]
print(freq)
    
for index, value in enumerate(a[1:]):
    if value != elements[-1]:
        elements.append(value)
        freq.append(1)
    elif value == elements[-1]:
        freq[-1] += 1

result = np.array(elements), np.array(freq)
result 
```
