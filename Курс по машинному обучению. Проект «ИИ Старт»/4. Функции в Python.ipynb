### **global** позволяет перезаписать значение глобальной переменной через локальную переменную функции

```python
ages = [18, 26, 13, 14, 15, 16]
sell = lambda x: True if x >= 18 else False
beer_shop = list(map(sell, ages))
beer_shop
```

```python
def almost_double_factorial(n):
    # YOUR CODE
    result = 1
    for i in range(1, n + 1):
        if i % 2 == 1:
            result *= i
            
    return result # YOUR CODE
```

```python
counter = -1


def calls_num():
    global counter
    counter += 1  # counter = 0 now
    # previously functuion has been called 0 times
    return counter
```

```python
first = 0
second = 0


def cmp_prev(val):
    # YOUR CODE
    global first  # Takes global values of variables
    global second
    
    result = val > first + second
    first = second  # Variables take new values
    second = val
    
    return result  # YOUR CODE
```
