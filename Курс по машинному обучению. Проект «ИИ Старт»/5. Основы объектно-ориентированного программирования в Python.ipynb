```python
class Electrical_device():
    def __init__(self, voltage, name):
        self.voltage = voltage
        self.name = name
        
        status = False

    def On(self):
        self.status = True
        print(" ", self.name)

    def Off(self):
        self.status = False
        print(" ", self.name)
```

```python
class Lamp(Electrical_device): #   
    brightness = 0

    def info(self):
        print(" ", self.status)
        print("E ", self.brightness)
```

```python
class Cleaner(Electrical_device) :
    type_of_cleaning = ""

    def cleaning(self):
        if self.type_of_cleaning == "":
            print ("")
        if self.type_of_cleaning == "":
            print ("")
```

```python
Cleaner1 = Cleaner(220, "Po ")
Lamp1 = Lamp(220, " ")
```

```python
Lamp1.On()
Lamp1.info()
```

```python
Cleaner1.type_of_cleaning = ""
Cleaner1.cleaning()
```

```python
class MyNum:
    #YOUR CODE HERE
    def __init__(self, number):
        self.number = number
        
    def __iadd__(self, other):
        self.number += other
        return self
    
    def get_val(self):
        return self.number
```

```python
class MyQueue:
    def __init__(self):
        #YOUR CODE HERE
        self.queue = []

    def push(self, val):
        #YOUR CODE HERE
        self.queue.append(val)

    def front(self):
        #YOUR CODE HERE
        return self.queue[0]
        
    def pop(self):
        #YOUR CODE HERE
        self.queue.pop(0)
```

```python
class Neuron:
    def __init__(self, w, f = lambda x: x):
        #YOUR CODE HERE
        self.w = w
        self.f = f
        self.x = None  #    forward  

    def forward(self, x):
        #YOUR CODE HERE
        self.x = x
        return self.f(sum([i * j for (i, j) in zip(self.w, self.x)]))

    def backlog(self):
        #YOUR CODE HERE
        return self.x
```

```python
###   ###
import math
import numpy as np

flag = True

w = [0.5, 1., -1., 2.]
failed = np.zeros(3)
passed = np.zeros(3)

test_groups = ['init', 
               'forward', 
               'backlog']
test_count = [1, 2, 2]

try:
    neuron = Neuron([0.5, 1., -1., 2.], 
                    lambda x: 1. / (1 + math.exp(x)))
    passed[0] = 1
except Exception as e:
    failed[0] = 1

x = [1, 2, 3, 4]

try:
    if abs(neuron.forward(x) - 0.0005527) > 1e-6:
        failed[1] += 1
    else:
        passed[1] += 1
except Exception as e:
    failed[1] += 1
    
try:
    if neuron.backlog() != x:
        failed[2] += 1
    else:
        passed[2] += 1
except Exception as e:
    failed[2] += 1    

y = [0, 0, 0, 0]

try:
    if abs(neuron.forward(y) - 0.5) > 1e-6:
        failed[1] += 1
    else:
        passed[1] += 1
except Exception as e:
    failed[1] += 1

try:
    if neuron.backlog() != y:
        failed[2] += 1
    else:
        passed[2] += 1
except Exception as e:
    failed[2] += 1
    
if (failed == 0).all():
    reply = 'OK!'
else:
    reply = 'Detailed info below.\n'
    reply += '----------------------------'
    reply += '\nTest passed by testgroup:\n'
    for i in np.arange(len(test_groups)):
        reply += f'Testgroup {i + 1} ({test_groups[i]}): '
        reply += f'passed {int(passed[i])} from {test_count[i]}\n'
print(reply)
```
