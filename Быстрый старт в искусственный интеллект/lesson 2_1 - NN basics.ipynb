**Нейрон** - это по-сути обобщенная логистическая регрессия. Обобщенная потому что функция активации может быть не только сигмоида, но и другие.

Внутри одного слоя связи между нейронами нет

### Входной слой нейронной сети содержит столько же нейронов, сколько признаков в данных (или пикселей на изображении)

Последний слой нейронной сети содержит ответ, чаще в виде вероятностей принадлежности к каждому из классов или ответ в виде значения в одном нейроне последнего слоя



**Нейронная сеть** - это композиция множества простых функций в нейронах (сложная функция)

Больше степеней свободы в НС дают возможность находить более сложные закономерности в данных

**Батч** - это кусок данных, которые НС прогоняет через себя для подсчета функции потерь и обновления весов
